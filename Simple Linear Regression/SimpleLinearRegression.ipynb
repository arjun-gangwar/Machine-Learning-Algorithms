```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_raw = pd.read_csv("data/train.txt", header=None)
data_raw.head()
```

```python
X = data_raw[0]
X.head()
```

```python
y = data_raw[1]
y.head()
```

```python
X = X[:,np.newaxis]
y = y[:,np.newaxis]
```

```python
X.shape
```

```python
y.shape
```

```python
X = np.hstack((np.ones((X.shape[0],1)), X))
X.shape
```

```python
initialTheta = np.array([[0],[0]])
print(initialTheta.shape)
```

```python
def gradientDescent(theta, X, y, alpha):
    J, h = computeCost(X, y, theta)
    print("j",J)
    m = X.shape[0]
    d = h - y
    t = np.dot(X.T,d)
    t = (alpha * t) / m
    theta = theta - t
    return theta, h
```

```python
def computeCost(X, y, theta):
    m = X.shape[0]
    h = np.dot(X,theta)
    J = np.power(h-y,2)
    J = np.sum(J, axis=0)
    J = J / (2*m)
    return J, h
```

```python
def train(theta, X, y, alpha, maxIter):
    for i in range(maxIter):
        theta, h = gradientDescent(theta, X, y, alpha)
    return theta
```

```python
def predict(p,theta):
    p = p/10000
    p = np.array([1,p])
    p = np.dot(p, theta)
    p = p * 10000
    return p
```

# Training

```python
alpha = 0.01
maxIter = 1500
finalTheta = train(initialTheta, X, y, alpha, maxIter);
```

```python
plt.figure(figsize=(12,9))
plt.scatter(data_raw[0], data_raw[1], c='r')
plt.xlabel('size')
plt.ylabel('price')
plt.plot(X[:,1],np.dot(X,finalTheta))
plt.show()
```
